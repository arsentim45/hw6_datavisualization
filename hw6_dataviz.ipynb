```python
import altair as alt
import pandas as pd
```

```python
# read excel
df = pd.read_excel('1.4_02.xlsx')
df.head()
```

```python
year = []
region = []
for i in range(len(df)):
    year.append(df['Регіон, рік'][i][-4:])
    region.append(df['Регіон, рік'][i][:-6])
```

```python
region
```

```python
df['Регіон'] = region
df['Рік'] = year
```

```python
df.head()
```

```python
df1 = df.melt(id_vars=["Регіон, рік", "Регіон", "Рік"], 
        var_name="Тип", 
        value_name="Значення")
```

```python
df1.head()
```

```python
side_histogram = alt.Chart(df).mark_bar().encode(
    y = alt.Y('Регіон, рік:N'),
    x = alt.X('сільське господарство:Q'),
    color='Рік:N',
)

side_histogram.properties(width = 950, height = 500, background = '#F9F9F9', padding = 25)
```

```python
alt.Chart(df1).mark_circle(size=100).encode(
    alt.X('Значення:Q'),
    y='Тип:O',
    color='Рік:N',
    facet=alt.Facet('Регіон:O', columns=2),
).properties(
    width=300,
    height=100,
)
```

```python
side_histogram = alt.Chart(df1).mark_bar().encode(
    y = alt.Y('Регіон:N'),
    x = alt.X('Значення:Q'),
    color='Рік:N',
    facet=alt.Facet('Тип:O', columns=1),
)

side_histogram.properties(width = 750, height = 500, background = '#F9F9F9', padding = 25)
```
